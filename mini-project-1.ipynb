```python
## when key is pressed 
whenBlock = ''
for block in blocks:
    if(blocks[block]['opcode'] == 'event_whenkeypressed'):
        whenBlock = block
        break
print(whenBlock)
```

## pseudocode for first set of blocks

```python
txtFile = open('pseudocode.txt','w')

txtFile.write('when_space_isClicked\n')
txtFile.write('  BEGIN\n')

currentBlock = blocks[whenBlock]['next']
while currentBlock != None:
    inputs = blocks[currentBlock]['inputs']
    opcode = blocks[currentBlock]['opcode']
    if opcode == 'motion_gotoxy':
        x = inputs['X'][-1][-1]
        y = inputs['Y'][-1][-1]
        txtFile.write('    go_to_x_{x}_y_{y}\n'.format(x=x,y=y))
    elif opcode == 'motion_changexby':
        dx = inputs['DX'][-1][-1]
        txtFile.write('    change_x_by_{dx}\n'.format(dx=dx))
    elif opcode == 'motion_movesteps':
        steps = inputs['STEPS'][-1][-1]
        txtFile.write('    move_{steps}_steps\n'.format(steps=steps))
    elif opcode == 'motion_turnright':
        degrees = inputs['DEGREES'][-1][-1]
        txtFile.write('    turnR_{degrees}_degrees\n'.format(degrees=degrees))
    elif opcode == 'motion_setx':
        x = inputs['X'][-1][-1]
        txtFile.write('    set_x_{x}\n'.format(x=x))
    elif opcode == 'motion_changeyby':
        dy = inputs['DY'][-1][-1]
        txtFile.write('    change_y_by_{dy}\n'.format(dy=dy))
    elif opcode == 'motion_sety':
        y = inputs['Y'][-1][-1]
        txtFile.write('    set_y_{y}\n'.format(y=y))

    currentBlock = blocks[currentBlock]['next']    
        
    
    
txtFile.write('  END\n')
txtFile.close()
```

## second set of blocks

```python
## when key is pressed 
whenBlock = ''
for block in blocks:
    if(blocks[block]['opcode'] == 'event_whenflagclicked'):
        whenBlock = block
        break
print(whenBlock)
```

```python
txtFile = open('pseudocode.txt','a')
txtFile.write('when_GF_isClicked\n')
txtFile.write('  BEGIN\n')

currentBlock = blocks[whenBlock]['next']
while currentBlock != None:
    inputs = blocks[currentBlock]['inputs']
    opcode = blocks[currentBlock]['opcode']
    
    if opcode == 'looks_say':
        msg = inputs['MESSAGE'][-1][-1]
        txtFile.write('    say_{msg}\n'.format(msg=msg))
    elif opcode == 'looks_sayforsecs':
        msg = inputs['MESSAGE'][-1][-1]
        sec = inputs['SECS'][-1][-1]
        txtFile.write('    say_{msg}_for_{sec}_seconds\n'.format(msg=msg,sec=sec))
    elif opcode == 'looks_think':
        msg = inputs['MESSAGE'][-1][-1]
        txtFile.write('    think_{msg}\n'.format(msg=msg))
    elif opcode == 'looks_thinkforsecs':
        msg = inputs['MESSAGE'][-1][-1]
        sec = inputs['SECS'][-1][-1]
        txtFile.write('    think_{msg}_for_{sec}_seconds\n'.format(msg=msg,sec=sec))
        
    currentBlock = blocks[currentBlock]['next']


txtFile.write('  END\n')
txtFile.close()
```

## if block

```python
ifBlock = ''
for block in blocks:
    if(blocks[block]['opcode'] == 'control_if'):
        ifBlock = block
        break
print(ifBlock)
```

```python
txtFile = open('pseudocode.txt','a')


for block in blocks:
    if blocks[block]['parent'] == ifBlock and blocks[block]['opcode'] == 'motion_movesteps':
        inputs = blocks[block]['inputs']
        steps = inputs['STEPS'][-1][-1]
        ## search for contains
        string1 = ''
        string2 = ''
        for block2 in blocks:
            if blocks[block2]['parent'] == ifBlock and blocks[block2]['opcode'] == 'operator_contains':
                inputs = blocks[block2]['inputs']
                string1 = inputs['STRING1'][-1][-1]
                string2 = inputs['STRING2'][-1][-1]
                
        txtFile.write('if_{string1}_contains_{string2}\n'.format(string1 = string1, string2 = string2))
        txtFile.write('  BEGIN\n')
        txtFile.write('    move_{steps}_steps\n'.format(steps=steps))
        #print(blocks[block]['inputs'])

txtFile.write('  END\n')
txtFile.close()
#operator_contains
```

## if else block

```python
ifElseBlock = ''
for block in blocks:
    if(blocks[block]['opcode'] == 'control_if_else'):
        ifElseBlock = block
        break
print(ifElseBlock)
```

```python
txtFile = open('pseudocode.txt','a')

for block in blocks:
    if blocks[block]['parent'] == ifBlock and blocks[block]['opcode'] == 'motion_movesteps':
        steps = blocks[block]['inputs']['STEPS'][-1][-1]
        ## search for contains
        for block2 in blocks:
            if blocks[block2]['parent'] == ifElseBlock and blocks[block2]['opcode'] == 'operator_contains':
                inputs = blocks[block2]['inputs']
                string1 = inputs['STRING1'][-1][-1]
                string2 = inputs['STRING2'][-1][-1]
        txtFile.write('if_{string1}_contains_{string2}\n'.format(string1 = string1, string2 = string2))
        txtFile.write('  BEGIN\n')
        txtFile.write('    move_{steps}_steps\n'.format(steps=steps))
        txtFile.write('  END\n')
        degrees = ''
        for block3 in blocks:
            if blocks[block3]['parent'] == ifElseBlock and blocks[block3]['opcode'] == 'motion_turnleft':
                inputs = blocks[block3]['inputs']
                #print(inputs)
                degrees = inputs['DEGREES'][-1][-1]
        txtFile.write('ELSE\n')
        txtFile.write('  BEGIN\n')
        txtFile.write('    turnL_{degrees}_degrees\n'.format(degrees = degrees))


txtFile.write('  END\n')
txtFile.close()
```

## repeat until block

```python
repeatUntil = ''
for block in blocks:
    if(blocks[block]['opcode'] == 'control_repeat_until'):
        repeatUntil = block
        break
print(repeatUntil)
```

```python
txtFile = open('pseudocode.txt','a')

for block in blocks:
    # handle key is presses
    if blocks[block]['parent'] == repeatUntil and blocks[block]['opcode'] == 'sensing_keypressed':
        inputs = blocks[block]['inputs']
        blockOfKey = inputs['KEY_OPTION'][-1]
        #search for the key
        key = ''
        for block2 in blocks:
            if block2 == blockOfKey:
                fields = blocks[block2]['fields']
                key = fields['KEY_OPTION'][0]
                txtFile.write('repeat_until_{key}_isClicked\n'.format(key = key))
                txtFile.write('  BEGIN\n')
                #txtFile.write('    move_{steps}_steps\n'.format(steps=steps))
        
        #handle the body of until
        steps = ''
        for block3 in blocks:
            if blocks[block3]['parent'] == repeatUntil and blocks[block3]['opcode'] == 'motion_movesteps':
                inputs = blocks[block3]['inputs']
                steps = inputs['STEPS'][-1][-1]
                txtFile.write('    move_{steps}_steps\n'.format(steps=steps))
            
        
txtFile.write('  END\n')
txtFile.close()
```

## repeat and wait blocks

```python
def searchForNameOfBlock(blockcode):
    for block in blocks:
        if block == blockcode:
            return blocks[block]['opcode']
    return -1
```

```python
def getAllBlocks(opcode):
    repeatBlocks = []
    for block in blocks:
        if blocks[block]['opcode'] == opcode:
            repeatBlocks.append(block)
    return repeatBlocks

repeatBlocks = getAllBlocks('control_repeat')
print(repeatBlocks)
```

```python
txtFile = open('pseudocode.txt','a')
steps = ''
times = ''
waitTime = ''
degrees = ''
##search for blocks inside repeat block and the next is wait block
for block in blocks:

    if blocks[block]['parent'] in repeatBlocks and blocks[block]['next'] in getAllBlocks('control_wait'):
        repeatBlock = blocks[block]['parent']
        #get times
        times = blocks[repeatBlock]['inputs']['TIMES'][-1][-1]
        # get steps
        inputs = blocks[block]['inputs']
        steps = inputs['STEPS'][-1][-1]
        #get wait times
        waitblock = blocks[block]['next']
        waitTime = blocks[waitblock]['inputs']['DURATION'][-1][-1]
        # get degrees of next turn
        turnBlock = blocks[repeatBlock]['next']
        degrees = blocks[turnBlock]['inputs']['DEGREES'][-1][-1]
        
        
        
txtFile.write('repeat_{times}_times\n'.format(times = times))
txtFile.write('  BEGIN\n')
txtFile.write('    move_{steps}_steps\n'.format(steps = steps))
txtFile.write('    wait_{waitTime}\n'.format(waitTime = waitTime))
txtFile.write('  END\n')
txtFile.write('turnR_{degrees}_degrees\n'.format(degrees = degrees))
txtFile.close()

```

## wait until

```python
txtFile = open('pseudocode.txt','a')
steps = ''
times = ''
key = ''
degrees = ''
##search for blocks inside repeat block and the next is wait block
for block in blocks:

    if blocks[block]['parent'] in repeatBlocks and blocks[block]['next'] in getAllBlocks('control_wait_until'):
        repeatBlock = blocks[block]['parent']
        #get times
        times = blocks[repeatBlock]['inputs']['TIMES'][-1][-1]
        # get steps
        inputs = blocks[block]['inputs']
        steps = inputs['STEPS'][-1][-1]
        #get condition of wait until
        waitblock = blocks[block]['next']
        keySense = blocks[waitblock]['inputs']['CONDITION'][-1]
        keyOptionBlock = blocks[keySense]['inputs']['KEY_OPTION'][-1]
        key = blocks[keyOptionBlock]['fields']['KEY_OPTION'][0]
        # get degrees of next turn
        turnBlock = blocks[repeatBlock]['next']
        degrees = blocks[turnBlock]['inputs']['DEGREES'][-1][-1]
        
txtFile.write('repeat_{times}_times\n'.format(times = times))
txtFile.write('  BEGIN\n')
txtFile.write('    move_{steps}_steps\n'.format(steps = steps))
txtFile.write('    wait_until_{key}_isClicked\n'.format(key = key))
txtFile.write('  END\n')
txtFile.write('turnR_{degrees}_degrees\n'.format(degrees = degrees))
txtFile.close()

       
```
